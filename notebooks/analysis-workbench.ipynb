```python
import plotly.express as px
from sklearn.datasets import load_digits
from umap import UMAP

digits = load_digits()

print(type(digits), digits)
print(type(digits.data), digits.data)
print(type(digits.target), digits.target)

umap_2d = UMAP(random_state=0)
umap_2d.fit(digits.data)

projections = umap_2d.transform(digits.data)

fig = px.scatter(
    projections, x=0, y=1,
    color=digits.target.astype(str), labels={'color': 'digit'}
)
fig.show()
```
